## Faça você mesmo #1
Utilize o código abaixo para plotar vetores em um plano

```python
vetores = np.array([[-2,1],[3,4], [1,1]])
origin = np.array([[0, 0, 0],[0, 0, 0]]) # origem

plt.axvline(x=0, c="black", label="x=0")
plt.axhline(y=0, c="black", label="y=0")
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.quiver(*origin, vetores[:,0], vetores[:,1], color=['r','b','g'], angles='xy', scale_units='xy', scale=1)

plt.show()
```

## **Faça você mesmo #4**
Execute o próximo bloco para plotar a função de cosseno

```python
x = np.arange(-2*np.pi,2*np.pi,0.1)#início, fim, passo
cos_x = np.cos(x)

fig, ax = plt.subplots()
ax.set_facecolor("black") #trocar a cor de fundo
ax.plot(x,cos_x)
plt.axvline(x=0, c="white", label="x=0")
plt.axhline(y=0, c="white", label="y=0")
plt.show()
```
