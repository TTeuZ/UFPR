## Visualizando o problema XOR
Dado a visualização gerada pelo gráfico abaixo, é possível utilizar um perceptron para separar pontos vermelhos e pontos azuis?

```python
negativo = np.random.uniform(1.2,0.7,(10,2))
positivo = np.array(negativo)

negativo[0:5, :] *= -1
positivo[0:5,0] *= -1
positivo[5:10,1] *= -1

plt.xlabel("x1")
plt.ylabel("x2")

plt.axvline(x=0, c="black", label="x=0")
plt.axhline(y=0, c="black", label="y=0")
plt.xlim(-2, 2)
plt.ylim(-2, 2)

plt.plot(negativo[:,0], negativo[:,1], 'o', color='red')
plt.plot(positivo[:,0], positivo[:,1], '+', color='blue')
```

## Faça Você Mesmo #1

Implemente a MLP da imagem abaixo usando o scikit-learn e perceptrons simples.
image.png



```python
negativo = np.random.uniform(-2,2,(10000,2))
positivo = np.array(negativo)

negativo[0:5000, :] *= -1
positivo[0:5000,0] *= -1
positivo[5000:10000,1] *= -1

inputs = np.concatenate([negativo, positivo])

# ----------------------------------------------------------

n_hidden_1 = Perceptron()
n_hidden_1.n_features_in_ = 2
n_hidden_1.classes_ = np.array([-1, 1])
n_hidden_1.intercept_ = np.array([0.5])
n_hidden_1.coef_ = np.array([[1, 1]])

n_hidden_2 = Perceptron()
n_hidden_2.n_features_in_ = 2
n_hidden_2.classes_ = np.array([-1, 1])
n_hidden_2.intercept_ = np.array([-1.5])
n_hidden_2.coef_ = np.array([[1, 1]])

n_output = Perceptron()
n_output.n_features_in_ = 2
n_output.classes_ = np.array([-1, 1])
n_output.intercept_ = np.array([-1])
n_output.coef_ = np.array([[0.7, -1.4]])

# ----------------------------------------------------------

h_out_1 = n_hidden_1.predict(inputs)
h_out_2 = n_hidden_2.predict(inputs)

n_inputs = np.column_stack((h_out_1, h_out_2))
outputs = n_output.predict(n_inputs)

# ----------------------------------------------------------

plt.xlabel("x1")
plt.ylabel("x2")
plt.axvline(x=0, c="black", label="x=0")
plt.axhline(y=0, c="black", label="y=0")
plt.xlim(-2, 2)
plt.ylim(-2, 2)

for i, input in enumerate(inputs):
  if outputs[i] >= 0:
    plt.plot(input[0], input[1], '+', color='blue')
  else:
    plt.plot(input[0], input[1], 'o', color='red')
```

Modifique o código abaixo para transformar os dados disponibilizados no gráfico em coordenadas polares. Utilize as equações
image.png



```python
r = np.random.uniform([[0], [2]], [[1], [3]], size=(2, 100))
angle = np.linspace(0, 2 * np.pi , 100 )


X = np.array([r[0] * np.cos(angle), r[1] * np.cos(angle)])
Y = np.array([r[0] * np.sin(angle), r[1] * np.sin(angle)])

plt.axis('equal')
plt.scatter(X[0], Y[0], c="red")
plt.scatter(X[1], Y[1], c="blue")
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()

R = np.sqrt(np.power(X, 2) + np.power(Y, 2))
O = np.arctan(Y / X)

plt.axis('equal')
plt.scatter(R[0], O[0], c="red")
plt.scatter(R[1], O[1], c="blue")
plt.show()
```
