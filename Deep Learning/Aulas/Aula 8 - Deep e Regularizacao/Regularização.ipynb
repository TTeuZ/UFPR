```python
from torchvision.transforms import v2
import matplotlib.pyplot as plt
from pathlib import Path
from PIL import Image
import torch

plt.rcParams["savefig.bbox"] = "tight"
torch.manual_seed(0)

# Trocar para o path certo no pc
orig_img1 = Image.open(Path("./imgs") / "2012-12-16_15_35_12.jpg")
orig_img2 = Image.open(Path("./imgs") / "2013-03-16_08_30_02.jpg")
display(orig_img1)
display(orig_img2)
```

```python
vflipper = v2.RandomHorizontalFlip(p=1.0)

transformed_img = vflipper(orig_img1)
display(orig_img1)
display(transformed_img)

transformed_img = vflipper(orig_img2)
display(orig_img2)
display(transformed_img)
```

```python
resized_imgs = [v2.Resize(size=size)(orig_img1) for size in (30, 50, 100)]
display(orig_img1)

for img in resized_imgs:
  display(img)
```

```python
center_crops = [v2.CenterCrop(size=size)(orig_img2) for size in (34, 50, 100)]
display(orig_img2)

for img in center_crops:
  display(img)
```

```python
perspective_transformer = v2.RandomPerspective(distortion_scale=0.6, p=1.0)
perspective_imgs = [perspective_transformer(orig_img1) for _ in range(4)]
display(orig_img1)

for img in perspective_imgs:
  display(img)
```

```python
jitter = v2.ColorJitter(brightness=.5, hue=.3)
jittered_imgs = [jitter(orig_img2) for _ in range(4)]
display(orig_img2)

for img in jittered_imgs:
  display(img)
```

```python
blurrer = v2.GaussianBlur(kernel_size=(5, 9), sigma=(0.1, 5.))
blurred_imgs = [blurrer(orig_img1) for _ in range(4)]
display(orig_img1)

for img in blurred_imgs:
  display(img)
```

```python
posterizer = v2.RandomPosterize(bits=2)
posterized_imgs = [posterizer(orig_img2) for _ in range(4)]
display(orig_img2)

for img in posterized_imgs:
  display(img)
```
